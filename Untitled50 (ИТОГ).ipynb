## 1) Загрузка и подготовка данных

В ручном режиме файл был очищен от некоторых строк, которые препятствовали корректной оценке данных. Далее перейдём к проверке данных на выбросы. При этом основной целью будет скорее не найти выбросы, а посмотреть с каким типом распределения данных я имею дело.
Проверять буду следующие данные: "НМЦК", "Количество поданных заявок" и "День недели даты окончания срока подачи заявок". Данные категории были выбраны по следующим причинам:
- "НМЦК" это один из ключевых показателей в закупках, в используемой выборке данных он является единственным финансовым критерием.
- "Количество поданных заявок" является ключевым показателем, поскольку именно отвечает за "успешность" соответственно важно узнать какое распределение дпнных по данному признаку.
- "День недели даты окончания срока подачи заявок". Проверяется именно дата "окончания" срока, а не "начала", поскольку про начало в законе о контрактной системе не сказано, что есть какие-то ограничения в размещении закупок, а на счёт окончания есть прямое указание на то, что окончание подачи заявок должно выпадать на рабочий день. Соответственно наличие в данных "субботы" или "воскресенья" может говорить о нарушении (технически у заказчика заблокированна возможности поставить ДОПЗ в выходной день) соответственно любая закупка с ДОПЗ в выходной день является результатом технических неполадок со стороны ЕИС и требует точечного анализа и "чистки".

Другие критерии нецелесообразно проверять, поскольку они либо имеют небольшое количество вариантов (например в 'Способ определения поставщика' есть только 3 варианта значений, а в месяцах - 12), либо могут иметь большое количество вариантов значений, как например в колонках с "ОКПД 2", но при этом данныйе значения не являются по своей сути числовой переменной (это классификатор).

#### Переходим к анализу выбросов

```python
print("\nАнализ выбросов в НМЦК:\n")

print("Анализ распределения НМЦК:")
nmck_stats = dfw['nmck'].describe()
print(nmck_stats)

# Коэффициент асимметрии
from scipy.stats import skew
skewness = skew(dfw['nmck'].dropna())
print(f"Коэффициент асимметрии: {skewness:.2f}")

if skewness > 1:
    print("Сильная правая асимметрия") 
    method = "IQR"
elif skewness < -1:
    print("Сильная левая асимметрия")
    method = "IQR"
else:
    print("Распределение близко к нормальному")
    method = "Z-score"

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(dfw['nmck'].dropna(), bins=50, alpha=0.7, edgecolor='black')
plt.title('Распределение НМЦК')
plt.xlabel('НМЦК')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(np.log1p(dfw['nmck'].dropna()), bins=50, alpha=0.7, edgecolor='black')
plt.title('Логарифмированное распределение НМЦК')
plt.xlabel('log(НМЦК + 1)')
plt.ylabel('Частота')
plt.tight_layout()
plt.show()

if method == "IQR":
    Q1 = dfw['nmck'].quantile(0.25)
    Q3 = dfw['nmck'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = dfw[(dfw['nmck'] < lower_bound) | (dfw['nmck'] > upper_bound)]
    print(f"Границы IQR: [{lower_bound:.2f}, {upper_bound:.2f}]")
else:
    from scipy import stats
    z_scores = np.abs(stats.zscore(dfw['nmck'].dropna()))
    outliers = dfw[z_scores > 3]
    print("Использован Z-score метод с порогом 3σ")

print(f"Выбросы в НМЦК: {len(outliers)} записей ({len(outliers)/len(dfw)*100:.2f}%)")
```

#### Не смотря на то, что имеется явная ассиметрия решаю не обрезать хвост. Мне будет достаточно информации о том, что уместнее будет использовать при дальнейшем анализе медианные значения, а не средние.

```python
print("\nАнализ выбросов в количестве поданных заявок (bids_cnt):\n")

print("Анализ распределения bids_cnt:")
bids_stats = dfw['bids_cnt'].describe()
print(bids_stats)

# Анализ асимметрии для дискретной переменной
skewness_bids = skew(dfw['bids_cnt'])
print(f"Коэффициент асимметрии: {skewness_bids:.2f}")

# Анализ модальности
mode_value = dfw['bids_cnt'].mode()[0]
print(f"Мода: {mode_value}")


if skewness_bids > 1.5:
    print("Сильная правая асимметрия")
    method_bids = "IQR"
elif dfw['bids_cnt'].nunique() <= 10:
    print("Мало уникальных значений")
    method_bids = "frequency"
else:
    print("Умеренная асимметрия")
    method_bids = "IQR"


plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
dfw['bids_cnt'].hist(bins=30, alpha=0.7, edgecolor='black')
plt.title('Распределение количества заявок')
plt.xlabel('Количество заявок')
plt.ylabel('Частота')

plt.subplot(1, 3, 2)
dfw['bids_cnt'].value_counts().sort_index().head(15).plot(kind='bar', alpha=0.7)
plt.title('Топ-15 значений по частоте')
plt.xlabel('Количество заявок')
plt.ylabel('Частота')

plt.subplot(1, 3, 3)
# Boxplot для визуализации выбросов
plt.boxplot(dfw['bids_cnt'].dropna())
plt.title('Boxplot количества заявок')
plt.ylabel('Количество заявок')

plt.tight_layout()
plt.show()


if method_bids == "IQR":
    Q1_bids = dfw['bids_cnt'].quantile(0.25)
    Q3_bids = dfw['bids_cnt'].quantile(0.75)
    IQR_bids = Q3_bids - Q1_bids
    lower_bound_bids = max(0, Q1_bids - 1.5 * IQR_bids)  # Не меньше 0
    upper_bound_bids = Q3_bids + 1.5 * IQR_bids
    
    outliers_bids = dfw[(dfw['bids_cnt'] < lower_bound_bids) | (dfw['bids_cnt'] > upper_bound_bids)]
    print(f"Границы IQR: [{lower_bound_bids:.2f}, {upper_bound_bids:.2f}]")
    
elif method_bids == "frequency":
    # Метод на основе частотности для дискретных данных
    freq_dist = dfw['bids_cnt'].value_counts(normalize=True)
    rare_values = freq_dist[freq_dist < 0.001]  # Значения с частотой < 0.1%
    outliers_bids = dfw[dfw['bids_cnt'].isin(rare_values.index)]
    print(f"Редкие значения (частота < 0.1%): {list(rare_values.index)}")

print(f"Выбросы в bids_cnt: {len(outliers_bids)} записей ({len(outliers_bids)/len(dfw)*100:.2f}%)")
```

#### По аналогии с данными по НМЦК решаю не обрезать хвост. Мне будет достаточно информации о том, что уместнее будет использовать при дальнейшем анализе медианные значения, а не средние.

##### Осталось проанализировать день недели ДОПЗ
Поскольку характер данных не числовой, а скорее категориальный буду использовать анализ частотности (посмотрим на 1% отклонений) и для наглядности визуализируем это распределение.

```python
print("\nАнализ дней недели окончания подачи заявок:\n")

dow_deadline_counts = dfw['dow_deadline'].value_counts(normalize=True)
print("Распределение по дням недели:")
print((dow_deadline_counts * 100).round(2))

total_records = len(dfw)
anomaly_threshold = 0.01

rare_categories = dow_deadline_counts[dow_deadline_counts < anomaly_threshold]
print(f"\nРедкие категории (менее {anomaly_threshold*100}%):")
for day, freq in rare_categories.items():
    count = int(freq * total_records)
    print(f"{day}: {count} записей ({freq*100:.2f}%)")

plt.figure(figsize=(10, 5))
dow_deadline_counts.sort_index().plot(kind='bar', alpha=0.7, edgecolor='black')
plt.title('Распределение закупок по дням недели')
plt.xlabel('День недели')
plt.ylabel('Доля закупок')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

valid_days = ['понедельник', 'вторник', 'среда', 'четверг', 'пятница', 'суббота', 'воскресенье']
invalid_days = dfw[~dfw['dow_deadline'].isin(valid_days)]

print(f"\nНекорректные значения в dow_deadline: {len(invalid_days)} записей")
```

Распределение по дням недели показывает явную "просадку" двух дней:
1) Воскресенье 11 записей (0,04%)
2) суббота: 87 записей (0.31%)

Поскольку закон о контрактной системе не допускает чтобы день окончания подачи заявок выпадал на выходной день, перейдём в исходный файл с данными чтобы точечно изучить все эти закупки.
Ручной анализ показал, что все закупки размещённые в субботу были размещены без нарушения закона о контрактной системе, поскольку в каждом таком случае такая "суббота" являлась рабочей согласно производственному календарю. Такие закупки исключать из анализа не следует, поскольку они хоть и являются редкими, но по своей сути не являются неправильными.
При проверке 11 закупок с ДОПЗ "воскресенье", было выявлено, что эти закупки были размещены одним и тем же заказчиком, в одну и туже дату с одним и тем же предметом закупки. Однако оставался вопрос каким образом всё-таки заказчик технически смог поставить не рабочий день (очевидно, что во всех остальных закупках фигурируют исключительно рабочие дни, кроме этих 11). Более глубокий анализ выявил, что во все эти закупки были внесены изменения, после которых заказчик продлял срок подачи заявок. Предполагается что именно этот фактор повлиял на то, что при указании выходного дня (именно не при первичном указании сроков, а при продлении) со стороны ЕИС не сработал блокирующий конроль, препядствующий размещению закупки. В данном случае можно предположить что возможно в этот период времени в ЕИС проводились технические работы (или что-то подобное) из-за чего возникла такая техническая "брежь", которой и воспользовался заказчик.

Исходя из вышесказанного было принято решение об исключении из анализа закупок с ДОПЗ "воскресенье"

```python
# Фильтр воскресений
before = len(dfw)
sunday_mask = dfw['dow_deadline'].astype(str).str.lower().str.strip().eq('воскресенье')
removed = int(sunday_mask.sum())
dfw = dfw.loc[~sunday_mask].reset_index(drop=True)
after = len(dfw)

print(f"Удалено строк (ДОПЗ = воскресенье): {removed}. Осталось: {after} из {before}")
```

##### Что касается влияния таких признаков, как НМЦК, месяцы, дни недели и коды ОКПД2, это требует отдельного рассмотрения.

#### Начну с рассмотрения влияния НМЦК и сразу проверю несколько гипотиз:
Исходя из увиденного в предыдущих выводах хочу проверить несколько гипотез:
- Чем больше НМЦК, тем больше вероятность что будет подано 2+ заявок
- Чем меньше НМЦК, тем больше вероятность, что будет подана 1 заявка
- Чем меньше НМЦК, тем больше вероятность, что будет подано 0 заявок

```python
print("\nСтатистический анализ и проверка гипотиз")

colors_comp = {'0': '#B22222', '1': '#FFD700', '2+': '#008000'}
category_order = ['0', '1', '2+']



if 'bids_category' not in dfw.columns:
    # Создаем только если не существует
    dfw['bids_category'] = dfw['bids_cnt'].apply(lambda x: '0' if x == 0 else '1' if x == 1 else '2+')


print("\nГипотезы по влиянию НМЦК на количество заявок:")
print("1. Чем больше НМЦК, тем больше вероятность что будет подано 2+ заявок")
print("2. Чем меньше НМЦК, тем больше вероятность, что будет подана 1 заявка") 
print("3. Чем меньше НМЦК, тем больше вероятность, что будет подано 0 заявок")


print("\nАнализ распределения НМЦК по категориям заявок:")


nmck_stats_by_bids = dfw.groupby('bids_category')['nmck'].describe()
print("\nОписательная статистика НМЦК по категориям заявок:")
print(nmck_stats_by_bids)


plt.figure(figsize=(15, 10))


plt.subplot(2, 2, 1)
sns.boxplot(data=dfw, x='bids_category', y='nmck', showfliers=False, 
            palette=colors_comp, order=category_order)
plt.yscale('log')
plt.title('Распределение НМЦК по категориям заявок\n(логарифмическая шкала)')
plt.xlabel('Категория заявок')
plt.ylabel('log(НМЦК)')


plt.subplot(2, 2, 2)
for category in category_order:
    subset = dfw[dfw['bids_category'] == category]
    plt.hist(np.log1p(subset['nmck']), alpha=0.3, label=f'{category} заявок', 
             bins=30, color=colors_comp[category], edgecolor=colors_comp[category], linewidth=0.5)
    
plt.title('Распределение НМЦК по категориям заявок')
plt.xlabel('log(НМЦК + 1)')
plt.ylabel('Частота')
plt.legend()


print("\nСтатистические тесты для проверки гипотез:")

# Тест Крускала-Уоллиса для сравнения медиан
from scipy.stats import kruskal


nmck_0 = dfw[dfw['bids_category'] == '0']['nmck']
nmck_1 = dfw[dfw['bids_category'] == '1']['nmck'] 
nmck_2plus = dfw[dfw['bids_category'] == '2+']['nmck']

# Тест Крускала-Уоллиса
h_stat, p_value = kruskal(nmck_0, nmck_1, nmck_2plus)
print(f"Тест Крускала-Уоллиса: H-статистика = {h_stat:.4f}, p-value = {p_value:.4f}")

if p_value < 0.05:
    print("Статистически значимые различия в распределении НМЦК между группами")
else:
    print("Нет статистически значимых различий в распределении НМЦК между группами")

# Попарные сравнения U-тестом Манна-Уитни
from scipy.stats import mannwhitneyu

print("\nПопарные сравнения (U-тест Манна-Уитни):")


stat_0_1, p_0_1 = mannwhitneyu(nmck_0, nmck_1, alternative='two-sided')
print(f"0 заявок vs 1 заявка: p-value = {p_0_1:.4f}")


stat_0_2, p_0_2 = mannwhitneyu(nmck_0, nmck_2plus, alternative='two-sided')
print(f"0 заявок vs 2+ заявок: p-value = {p_0_2:.4f}")


stat_1_2, p_1_2 = mannwhitneyu(nmck_1, nmck_2plus, alternative='two-sided')
print(f"1 заявка vs 2+ заявок: p-value = {p_1_2:.4f}")


print("\nСравнение средних и медиан НМЦК:")
comparison_df = pd.DataFrame({
    'Категория': ['0 заявок', '1 заявка', '2+ заявок'],
    'Среднее': [format_number(nmck_0.mean(), 0), format_number(nmck_1.mean(), 0), format_number(nmck_2plus.mean(), 0)],
    'Медиана': [format_number(nmck_0.median(), 0), format_number(nmck_1.median(), 0), format_number(nmck_2plus.median(), 0)],
    'Количество': [format_number(len(nmck_0)), format_number(len(nmck_1)), format_number(len(nmck_2plus))]
})

print(comparison_df)


plt.subplot(2, 2, 3)
categories = ['0 заявок', '1 заявка', '2+ заявок']
means = [nmck_0.mean(), nmck_1.mean(), nmck_2plus.mean()]
medians = [nmck_0.median(), nmck_1.median(), nmck_2plus.median()]

x = np.arange(len(categories))
width = 0.35

plt.bar(x - width/2, means, width, label='Среднее', alpha=0.8, color=[colors_comp['0'], colors_comp['1'], colors_comp['2+']])
plt.bar(x + width/2, medians, width, label='Медиана', alpha=0.8, color=[colors_comp['0'], colors_comp['1'], colors_comp['2+']], hatch='//')
plt.xlabel('Категория заявок')
plt.ylabel('НМЦК (руб)')
plt.title('Сравнение средних и медиан НМЦК')
plt.xticks(x, categories)
plt.legend()
plt.yscale('log')


from matplotlib.ticker import FuncFormatter
def y_fmt(x, pos):
    return format_number(x, 0)
plt.gca().yaxis.set_major_formatter(FuncFormatter(y_fmt))


plt.subplot(2, 2, 4)
# Разделим НМЦК на квартили для анализа вероятностей
dfw['nmck_quartile'] = pd.qcut(dfw['nmck'], q=4, labels=['Q1 (низкий)', 'Q2', 'Q3', 'Q4 (высокий)'])


prob_analysis = dfw.groupby('nmck_quartile')['bids_category'].value_counts(normalize=True).unstack().fillna(0)
prob_analysis = prob_analysis[category_order]  # Убедимся в правильном порядке столбцов


quartile_ranges = dfw.groupby('nmck_quartile')['nmck'].agg(['min', 'max']).round(2)
labels = []
for q in quartile_ranges.index:
    ranges = quartile_ranges.loc[q]
    labels.append(f"Q{list(quartile_ranges.index).index(q)+1}\n({format_number(ranges['min'], 0)}-{format_number(ranges['max'], 0)})")


ax = prob_analysis.plot(kind='bar', stacked=True, ax=plt.gca(), color=[colors_comp[cat] for cat in category_order])
plt.title('Вероятность категорий заявок по квартилям НМЦК')
plt.xlabel('Квартиль НМЦК')
plt.ylabel('Вероятность')
plt.legend(title='Категория заявок')
plt.xticks(range(len(labels)), labels, rotation=45, ha='right')


for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='center', fontsize=8)

plt.tight_layout()
plt.savefig(FIG / 'nmck_analysis.png', dpi=300, bbox_inches='tight')
plt.show()


print("\nПроверка гипотез:")

median_2plus = nmck_2plus.median()
median_1 = nmck_1.median()
median_0 = nmck_0.median()

print(f"\nГипотеза 1: Чем больше НМЦК, тем больше вероятность 2+ заявок")
print(f"Медиана НМЦК для 2+ заявок: {format_number(median_2plus, 0)} руб.")
print(f"Медиана НМЦК для 1 заявки: {format_number(median_1, 0)} руб.")
print(f"Медиана НМЦК для 0 заявок: {format_number(median_0, 0)} руб.")

if median_2plus > median_1 and p_1_2 < 0.05:
    print("Гипотеза подтверждена: большие НМЦК привлекают больше заявок")
else:
    print("Гипотеза не подтверждена")


print(f"\nГипотеза 2: Чем меньше НМЦК, тем больше вероятность 1 заявки")
if median_1 < median_2plus and p_1_2 < 0.05:
    print("Гипотеза подтверждена: средние НМЦК привлекают 1 заявку")
else:
    print("Гипотеза не подтверждена")


print(f"\nГипотеза 3: Чем меньше НМЦК, тем больше вероятность 0 заявок")
if median_0 < median_1 and p_0_1 < 0.05:
    print("Гипотеза подтверждена: маленькие НМЦК чаще остаются без заявок")
else:
    print("Гипотеза не подтверждена")


print("\nКорреляционный анализ:")
correlation = dfw['nmck'].corr(dfw['bids_cnt'])
print(f"Корреляция между НМЦК и количеством заявок: {correlation:.4f}")

if correlation > 0.1:
    print("Слабая положительная корреляция: с ростом НМЦК растет количество заявок")
elif correlation < -0.1:
    print("Слабая отрицательная корреляция: с ростом НМЦК уменьшается количество заявок")
else:
    print("Корреляция практически отсутствует")


nmck_analysis_results = {
    'hypothesis_1_supported': median_2plus > median_1 and p_1_2 < 0.05,
    'hypothesis_2_supported': median_1 < median_2plus and p_1_2 < 0.05,
    'hypothesis_3_supported': median_0 < median_1 and p_0_1 < 0.05,
    'correlation_nmck_bids': correlation,
    'kruskal_p_value': p_value,
    'median_0': median_0,
    'median_1': median_1,
    'median_2plus': median_2plus
}

print("\nКлючевые выводы по влиянию НМЦК:")

if nmck_analysis_results['hypothesis_1_supported']:
    print("Крупные закупки (высокий НМЦК) привлекают больше участников")
else:
    print("Размер закупки не является ключевым фактором для привлечения множества участников")

if nmck_analysis_results['hypothesis_3_supported']:
    print("Мелкие закупки часто остаются без заявок - требуют особого внимания")
else:
    print("Размер закупки не влияет значительно на отсутствие заявок")

print(f"Статистическая значимость различий: {'ДА' if p_value < 0.05 else 'НЕТ'}")
```

На последнем графике "вероятность категории заявок по квартилям НМЦК" как раз видно подтверждение гипотизы о том, что чем ниже цена закупки, тем больше вероятность того, что она не состоится.

```python
print("Анализ влияния месяцев на количество поданных заявок")

print("\nВлияние месяцев окончания подачи заявок на количество поданных заявок:")

month_deadline_stats = dfw.groupby('month_deadline').agg({
    'bids_cnt': ['count', 'mean', 'median'],
    'is_zero': 'mean',
    'is_two_plus': 'mean'
}).round(2)

month_deadline_stats.columns = ['Количество закупок', 'Среднее кол-во заявок', 'Медиана заявок', 
                               'Доля с 0 заявок', 'Доля с 2+ заявками']

month_order = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 
               'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
month_deadline_stats = month_deadline_stats.reindex(month_order)

print(month_deadline_stats)


best_month_deadline = month_deadline_stats['Доля с 2+ заявками'].idxmax()
worst_month_deadline = month_deadline_stats['Доля с 2+ заявками'].idxmin()
max_competition_deadline = month_deadline_stats['Доля с 2+ заявками'].max() * 100
min_competition_deadline = month_deadline_stats['Доля с 2+ заявками'].min() * 100

print(f"\nВыводы по месяцам окончания подачи:")
print(f"- Самый конкурентный месяц: {best_month_deadline} ({max_competition_deadline:.1f}% закупок с 2+ заявками)")
print(f"- Наименее конкурентный месяц: {worst_month_deadline} ({min_competition_deadline:.1f}% закупок с 2+ заявками)")
print(f"- Разница: {max_competition_deadline - min_competition_deadline:.1f} п.п.")


print("\nВлияние месяцев размещения извещения на количество поданных заявок:")

month_notice_stats = dfw.groupby('month_notice').agg({
    'bids_cnt': ['count', 'mean', 'median'],
    'is_zero': 'mean',
    'is_two_plus': 'mean'
}).round(2)

month_notice_stats.columns = ['Количество закупок', 'Среднее кол-во заявок', 'Медиана заявок', 
                             'Доля с 0 заявок', 'Доля с 2+ заявками']


month_order_notice = ['январь', 'февраль', 'март', 'апрель', 'май', 'июнь', 
                     'июль', 'август', 'сентябрь', 'октябрь', 'ноябрь', 'декабрь']
month_notice_stats = month_notice_stats.reindex(month_order_notice)

print(month_notice_stats)


best_month_notice = month_notice_stats['Доля с 2+ заявками'].idxmax()
worst_month_notice = month_notice_stats['Доля с 2+ заявками'].idxmin()
max_competition_notice = month_notice_stats['Доля с 2+ заявками'].max() * 100
min_competition_notice = month_notice_stats['Доля с 2+ заявками'].min() * 100

print(f"\nВыводы по месяцам размещения извещения:")
print(f"- Самый конкурентный месяц: {best_month_notice} ({max_competition_notice:.1f}% закупок с 2+ заявками)")
print(f"- Наименее конкурентный месяц: {worst_month_notice} ({min_competition_notice:.1f}% закупок с 2+ заявками)")
print(f"- Разница: {max_competition_notice - min_competition_notice:.1f} п.п.")

print("Общие выводы по сезонности:\n")

def detect_seasons(month_stats):
    """Автоматически определяет высокий и низкий сезоны на основе данных"""
    
    sorted_months = month_stats.sort_values('Доля с 2+ заявками', ascending=False)
        
    high_season = sorted_months.head(4).index.tolist()
        
    low_season = sorted_months.tail(2).index.tolist()
    
    return high_season, low_season


high_season_deadline, low_season_deadline = detect_seasons(month_deadline_stats)
high_season_avg_deadline = month_deadline_stats.loc[high_season_deadline]['Доля с 2+ заявками'].mean() * 100
low_season_avg_deadline = month_deadline_stats.loc[low_season_deadline]['Доля с 2+ заявками'].mean() * 100

print(f"По месяцам окончания подачи:")
print(f"- Высокий сезон ({', '.join(high_season_deadline)}): {high_season_avg_deadline:.1f}% закупок с конкуренцией")
print(f"- Низкий сезон ({', '.join(low_season_deadline)}): {low_season_avg_deadline:.1f}% закупок с конкуренцией")
print(f"- Сезонная разница: {high_season_avg_deadline - low_season_avg_deadline:.1f} п.п.")


high_season_notice, low_season_notice = detect_seasons(month_notice_stats)
high_season_avg_notice = month_notice_stats.loc[high_season_notice]['Доля с 2+ заявками'].mean() * 100
low_season_avg_notice = month_notice_stats.loc[low_season_notice]['Доля с 2+ заявками'].mean() * 100

print(f"\nПо месяцам размещения извещения:")
print(f"- Высокий сезон ({', '.join(high_season_notice)}): {high_season_avg_notice:.1f}% закупок с конкуренцией")
print(f"- Низкий сезон ({', '.join(low_season_notice)}): {low_season_avg_notice:.1f}% закупок с конкуренцией")
print(f"- Сезонная разница: {high_season_avg_notice - low_season_avg_notice:.1f} п.п.")


problem_month_deadline = month_deadline_stats['Доля с 0 заявок'].idxmax()
problem_rate_deadline = month_deadline_stats['Доля с 0 заявок'].max() * 100

problem_month_notice = month_notice_stats['Доля с 0 заявок'].idxmax()
problem_rate_notice = month_notice_stats['Доля с 0 заявок'].max() * 100

print(f"\nПроблемные месяцы (максимальная доля закупок без заявок):")
print(f"- По окончанию подачи: {problem_month_deadline} ({problem_rate_deadline:.1f}%)")
print(f"- По размещению извещения: {problem_month_notice} ({problem_rate_notice:.1f}%)")


plt.figure(figsize=(15, 10))


plt.subplot(2, 2, 1)
plt.plot(month_deadline_stats.index, month_deadline_stats['Среднее кол-во заявок'], marker='o')
plt.title('Среднее количество заявок по месяцам окончания подачи')
plt.xticks(rotation=45)
plt.ylabel('Среднее количество заявок')

plt.subplot(2, 2, 2)
plt.plot(month_deadline_stats.index, month_deadline_stats['Доля с 0 заявок'], marker='o', color='red')
plt.title('Доля закупок с 0 заявок по месяцам окончания подачи')
plt.xticks(rotation=45)
plt.ylabel('Доля закупок с 0 заявок')


plt.subplot(2, 2, 3)
plt.plot(month_notice_stats.index, month_notice_stats['Среднее кол-во заявок'], marker='o')
plt.title('Среднее количество заявок по месяцам размещения извещения')
plt.xticks(rotation=45)
plt.ylabel('Среднее количество заявок')

plt.subplot(2, 2, 4)
plt.plot(month_notice_stats.index, month_notice_stats['Доля с 0 заявок'], marker='o', color='red')
plt.title('Доля закупок с 0 заявок по месяцам размещения извещения')
plt.xticks(rotation=45)
plt.ylabel('Доля закупок с 0 заявок')

plt.tight_layout()
plt.savefig(FIG / 'month_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### Определю влияние оставшихся факторов на успешность закупки матрицой корреляций

```python
print("Итоговый анализ влияния факторов на количество поданных заявок")

print("\nКорреляционный анализ числовых признаков:")

corr_df = dfw[['nmck', 'bids_cnt', 'is_zero', 'is_one', 'is_two_plus', 
               'year_dl', 'month_num_dl', 'year_dn', 'month_num_dn']]

correlation_matrix = corr_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Матрица корреляций числовых признаков')
plt.tight_layout()
plt.savefig(FIG / 'correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

bids_corr = correlation_matrix['bids_cnt'].sort_values(ascending=False)
print("\nКорреляция признаков с количеством заявок:")
print(bids_corr)

print("Анализ и устранение мультиколлинеарности")

from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant  # Добавляем импорт
from sklearn.feature_selection import mutual_info_regression

def remove_multicollinearity(df, features, target_col='bids_cnt', vif_threshold=5, corr_threshold=0.7):
    """
    Функция для устранения мультиколлинеарности
    """
    print("Исходные признаки для анализа:", features)
    
    # 1. Анализ корреляций между признаками
    print("\nАнализ корреляций между признаками")
    corr_matrix = df[features].corr()
    
    # Находим сильно коррелирующие пары
    high_corr_pairs = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > corr_threshold:
                high_corr_pairs.append((
                    corr_matrix.columns[i], 
                    corr_matrix.columns[j], 
                    corr_matrix.iloc[i, j]
                ))
    
    if high_corr_pairs:
        print("Обнаружены сильно коррелирующие пары признаков:")
        for feat1, feat2, corr in sorted(high_corr_pairs, key=lambda x: abs(x[2]), reverse=True):
            print(f"  {feat1} <-> {feat2}: {corr:.3f}")
    else:
        print("Сильно коррелирующих пар не обнаружено")
    
    # 2. Расчет VIF (Variance Inflation Factor)
    print("\nАнализ VIF (VARIANCE INFLATION FACTOR):")
    X = df[features].fillna(df[features].median())
    
   
    X_with_const = add_constant(X)
    
    vif_data = pd.DataFrame()
    vif_data["feature"] = features
    vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i+1) 
                      for i in range(len(features))]
    
    print("VIF факторов:")
    print(vif_data.sort_values("VIF", ascending=False))
    

    print("\nАнализ важности признаков для целевой переменной:")
    
    # Mutual Information с целевой переменной
    if target_col in df.columns:
        X_filled = X.fillna(X.median())
        y = df[target_col].fillna(df[target_col].median())
        
        mi_scores = mutual_info_regression(X_filled, y, random_state=42)
        mi_data = pd.DataFrame({
            'feature': features,
            'MI_score': mi_scores
        }).sort_values('MI_score', ascending=False)
        
        print("Важность признаков (Mutual Information):")
        print(mi_data)
    else:
        mi_data = None
    
  
    print("\nОтбор признаков для исключения мультиколлинеарности:")
    
    features_to_keep = features.copy()
    features_to_remove = []
    
    # Удаляем один из каждой сильно коррелирующей пары (оставляем более важный)
    for feat1, feat2, corr in high_corr_pairs:
        if feat1 in features_to_keep and feat2 in features_to_keep:
            # Определяем какой признак оставить (по важности или произвольно)
            if mi_data is not None:
                # Оставляем признак с большей важностью
                mi1 = mi_data[mi_data['feature'] == feat1]['MI_score'].values[0]
                mi2 = mi_data[mi_data['feature'] == feat2]['MI_score'].values[0]
                
                if mi1 >= mi2:
                    features_to_keep.remove(feat2)
                    features_to_remove.append(feat2)
                    print(f"  Удаляем {feat2} (оставляем {feat1}, MI: {mi1:.3f} > {mi2:.3f})")
                else:
                    features_to_keep.remove(feat1)
                    features_to_remove.append(feat1)
                    print(f"  Удаляем {feat1} (оставляем {feat2}, MI: {mi2:.3f} > {mi1:.3f})")
            else:
                # Если нет целевой переменной, удаляем произвольно
                features_to_keep.remove(feat2)
                features_to_remove.append(feat2)
                print(f"  Удаляем {feat2} (сильная корреляция с {feat1}: {corr:.3f})")
    
    # Удаляем признаки с высоким VIF
    high_vif_features = vif_data[vif_data['VIF'] > vif_threshold]['feature'].tolist()
    for feature in high_vif_features:
        if feature in features_to_keep and feature not in features_to_remove:
            features_to_keep.remove(feature)
            features_to_remove.append(feature)
            print(f"  Удаляем {feature} (высокий VIF: {vif_data[vif_data['feature'] == feature]['VIF'].values[0]:.1f})")
    
    print(f"\nИтоговый список признаков для модели: {features_to_keep}")
    print(f"Удаленные признаки: {features_to_remove}")
    
    return features_to_keep, features_to_remove

print("\nАнализ мультиколлинеарности для числовых признаков:")
numeric_features = ['nmck', 'is_zero', 'is_one', 'is_two_plus', 
                   'year_dl', 'month_num_dl', 'year_dn', 'month_num_dn']

selected_features, removed_features = remove_multicollinearity(
    dfw, numeric_features, 'bids_cnt', vif_threshold=5, corr_threshold=0.7
)


print("\nСтатистический анализ значимости различий между группами:")

from scipy.stats import f_oneway, kruskal

# Проверка значимости различий между способами закупок
methods = [dfw[dfw['proc_method'] == method]['bids_cnt'] for method in dfw['proc_method'].unique()]
f_stat, p_value = f_oneway(*methods)
print(f"ANOVA test для способов закупок: F-statistic = {f_stat:.4f}, p-value = {p_value:.4f}")

if p_value < 0.05:
    print("Различия между способами закупок статистически значимы")
else:
    print("Различия между способами закупок не статистически значимы")

months_dl = [dfw[dfw['month_num_dl'] == month]['bids_cnt'] for month in dfw['month_num_dl'].unique() if len(dfw[dfw['month_num_dl'] == month]) > 10]
h_stat, p_value_month = kruskal(*months_dl)
print(f"Kruskal-Wallis test для месяцев окончания: H-statistic = {h_stat:.4f}, p-value = {p_value_month:.4f}")

if p_value_month < 0.05:
    print("Различия между месяцами окончания статистически значимы")
else:
    print("Различия между месяцами окончания не статистически значимы")

# 3.3 Анализ важности фRandom Forest
print("\nАнализ важности факторов с помощью Random Forest:")

# Подготовка данных для ML
ml_df = dfw.copy()


from sklearn.preprocessing import LabelEncoder

le_proc = LabelEncoder()
le_okpd2 = LabelEncoder()

ml_df['proc_method_encoded'] = le_proc.fit_transform(ml_df['proc_method'])
ml_df['okpd2_group_encoded'] = le_okpd2.fit_transform(ml_df['okpd2_group'])


final_features = selected_features + ['proc_method_encoded', 'okpd2_group_encoded']
print(f"\nФинальный набор признаков для модели: {final_features}")

X = ml_df[final_features]
y = ml_df['bids_cnt']

# Заполняем пропуски
X = X.fillna(X.median())

# Обучаем Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X, y)

feature_importance = pd.DataFrame({
    'feature': final_features,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nВажность признаков для предсказания количества заявок:")
print(feature_importance)


plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance, x='importance', y='feature', palette='viridis')
plt.title('Важность признаков (после устранения мультиколлинеарности)')
plt.xlabel('Важность')
plt.tight_layout()
plt.savefig(FIG / 'feature_importance_cleaned.png', dpi=300, bbox_inches='tight')
plt.show()


print("\nПроверка мультиколлинеарности после отбора признаков:")
final_corr_matrix = ml_df[final_features].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(final_corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Матрица корреляций после устранения мультиколлинеарности')
plt.tight_layout()
plt.savefig(FIG / 'correlation_matrix_cleaned.png', dpi=300, bbox_inches='tight')
plt.show()


print("Сводные выводы о влиянии факторов на количество поданных заявок")


print("\nФакторы, влияющие в большей степени:")


top_factors = feature_importance.head(3)
for i, row in top_factors.iterrows():
    feature_name = row['feature']
    importance = row['importance']
    
    if feature_name == 'nmck':
        print(f"1. НМЦК (начальная максимальная цена контракта) - важность: {importance:.3f}")
        print("   • Наиболее значимый фактор")
        print("   • Более высокие цены могут привлекать больше участников")
        
    elif feature_name == 'okpd2_group_encoded':
        print(f"2. Группа ОКПД2 - важность: {importance:.3f}")
        print("   • Второй по значимости фактор")
        print("   • Разные категории товаров/услуг привлекают разное количество участников")
        
    elif feature_name == 'proc_method_encoded':
        print(f"3. Способ закупки - важность: {importance:.3f}")
        print("   • Третий по значимости фактор")
        print("   • Разные процедуры имеют разную привлекательность для участников")


feature_importance.to_csv(CSV / 'feature_importance_cleaned.csv', encoding='utf-8-sig')
```

##### После того, как увидели, что на втором месте по влиянию стоит признак "ОКПД2", посмотрим какие коды ОКПД2 наиболее часто остаются без заявок или только с 1 заявкой. Для этого сделаем 2 топа (15).

*Определяем топ сразу по двум показателям (и по количеству и по сумме), отбираем топ ОКПД2 в которых оба ранга (по количеству и по сумме) занимают высокую позицию.*

```python
df_all_years = dfw[dfw['year_dl'].isin([2023, 2024, 2025])]

print("ПРОБЛЕМНЫЕ ОКПД2")

def shorten_name(name, max_words=5):
    if isinstance(name, str):
        words = name.split()
        if len(words) > max_words:
            return ' '.join(words[:max_words]) + '...'
        else:
            return name
    else:
        return "не указано"

# Таблица 1: Закупки без заявок
zero_all = df_all_years[df_all_years['is_zero'] == 1].groupby('okpd2_group_code').agg({
    'nmck': ['sum', 'count', 'median'],
    'okpd2_group': 'first'
}).round(2)
zero_all.columns = ['Совокупная_НМЦК', 'Количество_закупок', 'Медианная_НМЦК', 'Наименование']


zero_all['Ранг_по_сумме'] = zero_all['Совокупная_НМЦК'].rank(ascending=False, method='min').astype(int)
zero_all['Ранг_по_количеству'] = zero_all['Количество_закупок'].rank(ascending=False, method='min').astype(int)
zero_all['Средний_ранг'] = (zero_all['Ранг_по_сумме'] + zero_all['Ранг_по_количеству']) / 2


zero_combined = zero_all.sort_values('Средний_ранг').head(15)

print("\nТОП-15 ОКПД2 БЕЗ ЗАЯВОК С ВЫСОКИМИ РАНГАМИ ПО ОБОИМ ПОКАЗАТЕЛЯМ")
print("(ранги рассчитаны среди всех закупок без заявок за 2023-2025 годы)")
table_data = []
for i, (okpd2_code, row) in enumerate(zero_combined.iterrows(), 1):
    table_data.append([
        i,
        okpd2_code,
        shorten_name(row['Наименование']),
        format_number(row['Совокупная_НМЦК'], 0),
        int(row['Количество_закупок']),
        format_number(row['Медианная_НМЦК'], 0),
        row['Ранг_по_сумме'],
        row['Ранг_по_количеству'],
        round(row['Средний_ранг'], 1)
    ])

headers = ["№", "ОКПД2", "Наименование", "Сумма", "Кол-во", "Медиана", "Ранг сумм.", "Ранг кол-во", "Ср.ранг"]
print(tabulate(table_data, headers=headers, tablefmt="plain", stralign="left", numalign="left"))

# Таблица 2: Закупки с одной заявкой
one_all = df_all_years[df_all_years['is_one'] == 1].groupby('okpd2_group_code').agg({
    'nmck': ['sum', 'count', 'median'],
    'okpd2_group': 'first'
}).round(2)
one_all.columns = ['Совокупная_НМЦК', 'Количество_закупок', 'Медианная_НМЦК', 'Наименование']

# Добавляем ранги (среди всех закупок с одной заявкой)
one_all['Ранг_по_сумме'] = one_all['Совокупная_НМЦК'].rank(ascending=False, method='min').astype(int)
one_all['Ранг_по_количеству'] = one_all['Количество_закупок'].rank(ascending=False, method='min').astype(int)
one_all['Средний_ранг'] = (one_all['Ранг_по_сумме'] + one_all['Ранг_по_количеству']) / 2

# Отбираем ТОП-15 по среднему рангу
one_combined = one_all.sort_values('Средний_ранг').head(15)

print("\nТОП-15 ОКПД2 С ОДНОЙ ЗАЯВКОЙ С ВЫСОКИМИ РАНГАМИ ПО ОБОИМ ПОКАЗАТЕЛЯМ")
print("(ранги рассчитаны среди всех закупок с одной заявкой за 2023-2025 годы)")
table_data = []
for i, (okpd2_code, row) in enumerate(one_combined.iterrows(), 1):
    table_data.append([
        i,
        okpd2_code,
        shorten_name(row['Наименование']),
        format_number(row['Совокупная_НМЦК'], 0),
        int(row['Количество_закупок']),
        format_number(row['Медианная_НМЦК'], 0),
        row['Ранг_по_сумме'],
        row['Ранг_по_количеству'],
        round(row['Средний_ранг'], 1)
    ])

headers = ["№", "ОКПД2", "Наименование", "Сумма", "Кол-во", "Медиана", "Ранг сумм.", "Ранг кол-во", "Ср.ранг"]
print(tabulate(table_data, headers=headers, tablefmt="plain", stralign="left", numalign="left"))

print(f"\n*Всего закупок без заявок: {len(zero_all)} ОКПД2 категорий")
print(f"*Всего закупок с одной заявкой: {len(one_all)} ОКПД2 категорий")
print("*Сортировка по среднему рангу выявляет категории с высокими рангами как по сумме, так и по количеству")
```

```python
print("Среднестатистический портрет закупки\n")

def create_portrait(bid_type, label):
    """Создает портрет закупки для заданного типа"""
    subset = dfw[dfw[bid_type] == 1]
    
    if len(subset) == 0:
        return ["Нет данных"] * 5
    
    
    most_common_method = subset['proc_method'].mode()
    method = most_common_method[0] if not most_common_method.empty else "нет данных"
    
    
    median_nmck = subset['nmck'].median()
    
    
    most_common_okpd = subset['okpd2_group_code'].mode()
    okpd_code = most_common_okpd[0] if not most_common_okpd.empty else "нет данных"
    okpd_name = subset.loc[subset['okpd2_group_code'] == okpd_code, 'okpd2_group'].iloc[0] if not most_common_okpd.empty else "нет данных"
    
    
    most_common_dow = subset['dow_deadline'].mode()
    dow = most_common_dow[0] if not most_common_dow.empty else "нет данных"
    short_dow = {'понедельник': 'ПН', 'вторник': 'ВТ', 'среда': 'СР', 'четверг': 'ЧТ', 
                 'пятница': 'ПТ', 'суббота': 'СБ', 'воскресенье': 'ВС'}.get(dow, dow)
    
    
    most_common_month = subset['month_num_dl'].mode()
    month_num = most_common_month[0] if not most_common_month.empty else "нет данных"
    month_names = {1: 'январе', 2: 'феврале', 3: 'марте', 4: 'апреле', 5: 'мае', 6: 'июне',
                   7: 'июле', 8: 'августе', 9: 'сентябре', 10: 'октябре', 11: 'ноябре', 12: 'декабре'}
    month = month_names.get(month_num, 'нет данных')
    
    return [method, f"{format_number(median_nmck, 0)}", okpd_code, okpd_name, short_dow, month]


for bid_type, label in [('is_zero', '0 заявок'), 
                        ('is_one', '1 заявкой'), 
                        ('is_two_plus', '2+ заявками')]:
    
    subset = dfw[dfw[bid_type] == 1]
    
    if len(subset) > 0:
        method = subset['proc_method'].mode()[0]
        median_nmck = format_number(subset['nmck'].median(), 0)
        okpd_code = subset['okpd2_group_code'].mode()[0]
        okpd_name = subset.loc[subset['okpd2_group_code'] == okpd_code, 'okpd2_group'].iloc[0]
        dow = subset['dow_deadline'].mode()[0]
        short_dow = {'понедельник': 'понедельник', 'вторник': 'вторник', 'среда': 'среду', 'четверг': 'четверг', 
                     'пятница': 'пятницу', 'суббота': 'субботу', 'воскресенье': 'воскресенье'}.get(dow, dow)
        
        print(f"Среднестатистическая закупка с {label}:")
        print(f"   • Это {method.lower()} с медианной НМЦК {median_nmck} руб.")
        print(f"   • Код ОКПД2: {okpd_code} ({okpd_name})")
        print(f"   • Чаще всего проходит во {short_dow}")
        print()
```

#### 3) Вести просветительскую работу и с заказчиками и с поставщиками о проведении закупки способом "запрос котировок в электронной форме", поскольку у данного способа самая болшая доля несостоявшихся закупок. Выявление областей (по ОКПД") с низкой или отсутствующей конкуренцией позволит более точно обращаться к конкретным группам поставщиков, которые потенциально могут быть за интересованны в участии в закупке с тем или иным видом деятельности.
