#### Переходим к анализу выбросов

Выбор метода для анализа **выбросов в НМЦК** (метод будет выбираться в зависимости от асимметрии):

* **IQR метод** используется при сильной асимметрии (|skewness| > 1), потому что он более устойчив к выбросам (данный метод основан на разделении данных на квартили, и всё, что находится за пределами 1.5 × IQR от первого и третьего квартилей, считается выбросами)

  или

* **Z-score метод** используется при нормальном распределении, так как предполагает симметричность данных (данный метод основан на сравнении каждого значения со средним. Любое значение, которое будет отличаться от среднего на 3 и более стандартных отклонений, будет считаться выбросом)

```python
print("\nАнализ выбросов в НМЦК:\n")

print("Анализ распределения НМЦК:")
nmck_stats = dfw['nmck'].describe()
print(nmck_stats)

# Коэффициент асимметрии
from scipy.stats import skew
skewness = skew(dfw['nmck'].dropna())
print(f"Коэффициент асимметрии: {skewness:.2f}")

if skewness > 1:
    print("Сильная правая асимметрия") 
    method = "IQR"
elif skewness < -1:
    print("Сильная левая асимметрия")
    method = "IQR"
else:
    print("Распределение близко к нормальному")
    method = "Z-score"

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(dfw['nmck'].dropna(), bins=50, alpha=0.7, edgecolor='black')
plt.title('Распределение НМЦК')
plt.xlabel('НМЦК')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(np.log1p(dfw['nmck'].dropna()), bins=50, alpha=0.7, edgecolor='black')
plt.title('Логарифмированное распределение НМЦК')
plt.xlabel('log(НМЦК + 1)')
plt.ylabel('Частота')
plt.tight_layout()
plt.show()

if method == "IQR":
    Q1 = dfw['nmck'].quantile(0.25)
    Q3 = dfw['nmck'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = dfw[(dfw['nmck'] < lower_bound) | (dfw['nmck'] > upper_bound)]
    print(f"Границы IQR: [{format_number(lower_bound, 2)}, {format_number(upper_bound, 2)}]")
else:
    z_scores = np.abs(stats.zscore(dfw['nmck'].dropna()))
    outliers = dfw[z_scores > 3] 
    print("Использован Z-score метод с порогом 3σ")

print(f"Выбросы в НМЦК: {format_number(len(outliers))} записей ({len(outliers)/len(dfw)*100:.2f}%)")
```

Несмотря на то, что имеется явная асимметрия, решаю **не обрезать хвост**. Выбросы в НМЦК отражают крупные закупки, которые могут иметь другую динамику конкурентности. Обрезав хвост, можно упустить важные паттерны.

Поскольку **количество поданных заявок** является целевой переменной, очищать данные от выбросов тут не нужно.
В данном блоке посмотрим, какое распределение и асимметрия у данных.

```python
print("Анализ распределения в количестве заявок:")
bids_stats = dfw['bids_cnt'].describe()
print(bids_stats)

skewness_bids = stats.skew(dfw['bids_cnt']) 
print(f"Коэффициент асимметрии: {skewness_bids:.2f}")


mode_value = dfw['bids_cnt'].mode()[0]
print(f"Мода: {mode_value}")

if skewness_bids > 1.5:
    print("Сильная правая асимметрия")
    method_bids = "IQR"
elif dfw['bids_cnt'].nunique() <= 10:
    print("Мало уникальных значений")
    method_bids = "frequency"
else:
    print("Умеренная асимметрия")
    method_bids = "IQR"

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
dfw['bids_cnt'].hist(bins=30, alpha=0.7, edgecolor='black')
plt.title('Распределение количества заявок')
plt.xlabel('Количество заявок')
plt.ylabel('Частота')

plt.subplot(1, 3, 2)
dfw['bids_cnt'].value_counts().sort_index().head(15).plot(kind='bar', alpha=0.7)
plt.title('Топ-15 значений по частоте')
plt.xlabel('Количество заявок')
plt.ylabel('Частота')

plt.subplot(1, 3, 3)

plt.boxplot(dfw['bids_cnt'].dropna())
plt.title('Boxplot количества заявок')
plt.ylabel('Количество заявок')

plt.tight_layout()
plt.show()

if method_bids == "IQR":
    Q1_bids = dfw['bids_cnt'].quantile(0.25)
    Q3_bids = dfw['bids_cnt'].quantile(0.75)
    IQR_bids = Q3_bids - Q1_bids
    lower_bound_bids = max(0, Q1_bids - 1.5 * IQR_bids)
    upper_bound_bids = Q3_bids + 1.5 * IQR_bids
    
    outliers_bids = dfw[(dfw['bids_cnt'] < lower_bound_bids) | (dfw['bids_cnt'] > upper_bound_bids)]
    print(f"Границы IQR: [{format_number(lower_bound_bids, 2)}, {format_number(upper_bound_bids, 2)}]")
    
elif method_bids == "frequency":

    freq_dist = dfw['bids_cnt'].value_counts(normalize=True)
    rare_values = freq_dist[freq_dist < 0.001]
    outliers_bids = dfw[dfw['bids_cnt'].isin(rare_values.index)]
    print(f"Редкие значения (частота < 0.1%): {list(rare_values.index)}")

print(f"Выбросы в bids_cnt: {format_number(len(outliers_bids))} записей ({len(outliers_bids)/len(dfw)*100:.2f}%)")
```

Из анализа видим, что имеется достаточно сильная правосторонняя асимметрия (2,57), большинство значений сосредоточено в начале, что отлично видно на графике и говорит о том, что **большинство закупок имеет небольшое количество заявок** (это подтверждается значением медианы (1 заявка) и моды (1 заявка)). Boxplot как раз показывает, что медиана очень близка к границе первого квартиля, а выбросы начинаются с 7 заявок и простираются до 40. Таким образом, закупки с 7+ заявками являются закупками с аномально высокой конкуренцией и требуют детального анализа (всего таких закупок 1 634 или 5,78% от общего объема). Изучив данные закупки, возможно, удастся рассмотреть какие-то инсайты, которые позволят увеличить количество подаваемых заявок.

Распределение по дням недели показывает явную "просадку" двух дней:
1) Воскресенье 11 записей (0,04%)
2) суббота: 87 записей (0.31%)

Поскольку закон о контрактной системе не допускает чтобы день окончания подачи заявок выпадал на выходной день, перейдём в исходный файл с данными чтобы точечно изучить все эти закупки.
Ручной анализ показал, что все закупки размещённые в субботу были размещены без нарушения закона о контрактной системе, поскольку в каждом таком случае такая "суббота" являлась рабочей согласно производственному календарю. Такие закупки исключать из анализа не следует, поскольку они хоть и являются редкими, но по своей сути не являются неправильными.
При проверке 11 закупок с ДОПЗ "воскресенье", было выявлено, что эти закупки были размещены одним и тем же заказчиком, в одну и туже дату с одним и тем же предметом закупки. Однако оставался вопрос каким образом всё-таки заказчик технически смог поставить не рабочий день для даты начала подачи заявок (очевидно, что во всех остальных закупках фигурируют исключительно рабочие дни, кроме этих 11). Более глубокий анализ выявил, что во все эти закупки были внесены изменения, после которых заказчик продлял срок подачи заявок. Предполагается что именно этот фактор повлиял на то, что при указании выходного дня (именно не при первичном указании сроков, а при продлении) со стороны ЕИС не сработал блокирующий конроль, препядствующий размещению закупки. В данном случае можно предположить что возможно в этот период времени в ЕИС проводились технические работы (или что-то подобное) из-за чего возникла такая техническая "брежь", которой и воспользовался заказчик.

Исходя из вышесказанного было принято решение об исключении из анализа закупок с ДОПЗ "воскресенье"

```python
# Фильтр воскресений
before = len(dfw)
sunday_mask = dfw['dow_deadline'].astype(str).str.lower().str.strip().eq('воскресенье')
removed = int(sunday_mask.sum())
dfw = dfw.loc[~sunday_mask].reset_index(drop=True)
after = len(dfw)

print(f"Удалено строк (ДОПЗ = воскресенье): {format_number(removed)}. Осталось: {format_number(after)} из {format_number(before)}")
```

#### После того, как данные были успешно загружены и обработаны посмотрим общую информацию о закупках.

Сначала представим общую информацию в текстовой форме, а затем для удобства восприятия в визуальном формате

```python
print(f"Общая информация о закупках\n")
print(f"Общее количество закупок: {format_number(len(dfw))} ")
print(f"Общая сумма НМЦК: {format_number(dfw['nmck'].sum(), 0)} руб.")
print(f"Период (по ДОПЗ): {dfw['date_deadline'].min().strftime('%d.%m.%Y')} - {dfw['date_deadline'].max().strftime('%d.%m.%Y')}")
print(f"Охват: {format_number((dfw['date_deadline'].max() - dfw['date_deadline'].min()).days)} дней")

print("\nРаспределение НМЦК:")
nmck_stats = dfw['nmck'].describe()
print(f"Среднее: {format_number(nmck_stats['mean'], 0)} руб.")
print(f"Медиана: {format_number(nmck_stats['50%'], 0)} руб.")
print(f"Минимум: {format_number(nmck_stats['min'], 0)} руб.")
print(f"Максимум: {format_number(nmck_stats['max'], 0)} руб.")

print("\nКонкурентность закупок:")

competition_count = {
    'Нет заявок': dfw['is_zero'].sum(),
    'Одна заявка': dfw['is_one'].sum(),
    '2+ заявок': dfw['is_two_plus'].sum()
}

competition_sum = {
    'Нет заявок': dfw[dfw['is_zero'] == 1]['nmck'].sum(),
    'Одна заявка': dfw[dfw['is_one'] == 1]['nmck'].sum(),
    '2+ заявок': dfw[dfw['is_two_plus'] == 1]['nmck'].sum()
}

total_count = len(dfw)
total_sum = dfw['nmck'].sum()

table_data = []
for name in competition_count.keys():
    count = competition_count[name]
    sum_nmck = competition_sum[name]
    pct_count = (count / total_count) * 100
    pct_sum = (sum_nmck / total_sum) * 100
    formatted_count = format_number(count)
    formatted_sum = format_number(sum_nmck, 0)
    
    table_data.append([
        name,
        formatted_count,
        f"{pct_count:.1f}%",
        formatted_sum,
        f"{pct_sum:.1f}%"
    ])

headers = ["Тип конкурентности", "Кол-во", "%", "Сумма НМЦК", "%"]
print(tabulate(table_data, headers=headers, tablefmt="plain"))

print("\nРаспределение закупок по способам:")

total_sum = dfw['nmck'].sum()  # Общая сумма НМЦК

table_data = []

for method, count in dfw['proc_method'].value_counts().items():
    pct_count = (count / len(dfw)) * 100
    method_sum = dfw[dfw['proc_method'] == method]['nmck'].sum()
    pct_sum = (method_sum / total_sum) * 100
    
    table_data.append([
        method,
        format_number(count),
        f"{pct_count:.1f}%",
        format_number(method_sum, 0),
        f"{pct_sum:.1f}%"
    ])


headers = ["Способ закупки", "Кол-во", "%", "Сумма НМЦК", "%"]
print(tabulate(table_data, headers=headers, tablefmt="plain"))

print("\nРаспределение конкуренции по способам закупки:")

total_sum = dfw['nmck'].sum()

table_data = []

for method, count in dfw['proc_method'].value_counts().items():
    method_data = dfw[dfw['proc_method'] == method]
    method_sum = method_data['nmck'].sum()
    pct_sum = (method_sum / total_sum) * 100
    
    
    zero_count = method_data['is_zero'].sum()
    one_count = method_data['is_one'].sum()
    two_plus_count = method_data['is_two_plus'].sum()
    
    zero_pct = (zero_count / count) * 100
    one_pct = (one_count / count) * 100
    two_plus_pct = (two_plus_count / count) * 100
    
    table_data.append([
        method,
        format_number(count),
        f"{pct_sum:.1f}%",
        # Конкурентность: 0 заявок
        format_number(zero_count),
        f"{zero_pct:.1f}%",
        # Конкурентность: 1 заявка
        format_number(one_count),
        f"{one_pct:.1f}%",
        # Конкурентность: 2+ заявок
        format_number(two_plus_count),
        f"{two_plus_pct:.1f}%"
    ])


headers = [
    "Способ закупки", 
    "Всего", 
    "% суммы", 
    "0 заявок", "%", 
    "1 заявка", "%", 
    "2+ заявок", "%"
]

print(tabulate(table_data, headers=headers, tablefmt="plain"))

def find_max_values(table_data):
    """Находит строки с максимальными значениями количества и суммы"""
    max_count_idx = 0
    max_sum_idx = 0
    max_count_val = 0
    max_sum_val = 0
    
    for i, row in enumerate(table_data):
        try:
            
            count_val = float(row[2].replace('%', '')) 
            sum_val = float(row[4].replace('%', ''))  
            
            if count_val > max_count_val:
                max_count_val = count_val
                max_count_idx = i
            if sum_val > max_sum_val:
                max_sum_val = sum_val
                max_sum_idx = i
        except (ValueError, IndexError):
            continue
    
    return max_count_idx, max_sum_idx


print("\nРаспределение по дням недели окончания подачи заявок:")

dow_deadline_counts = dfw['dow_deadline'].value_counts()
table_data_deadline = []

for day, count in dow_deadline_counts.items():
    pct_count = (count / len(dfw)) * 100
    day_sum = dfw[dfw['dow_deadline'] == day]['nmck'].sum()
    pct_sum = (day_sum / total_sum) * 100
    
    table_data_deadline.append([
        day,
        format_number(count),
        f"{pct_count:.1f}%",
        format_number(day_sum, 0),
        f"{pct_sum:.1f}%"
    ])

headers = ["День недели (окончание)", "Кол-во", "%", "Сумма НМЦК", "%"]
print(tabulate(table_data_deadline, headers=headers, tablefmt="plain"))


max_count_idx, max_sum_idx = find_max_values(table_data_deadline)
print(f"Наибольшее количество закупок: {table_data_deadline[max_count_idx][0]} - {table_data_deadline[max_count_idx][2]}")
print(f"Наибольшая сумма НМЦК: {table_data_deadline[max_sum_idx][0]} - {table_data_deadline[max_sum_idx][4]}")


print("\nРаспределение по дням недели начала подачи заявок:")

dow_notice_counts = dfw['dow_notice'].value_counts()
table_data_notice = []

for day, count in dow_notice_counts.items():
    pct_count = (count / len(dfw)) * 100
    day_sum = dfw[dfw['dow_notice'] == day]['nmck'].sum()
    pct_sum = (day_sum / total_sum) * 100
    
    table_data_notice.append([
        day,
        format_number(count),
        f"{pct_count:.1f}%",
        format_number(day_sum, 0),
        f"{pct_sum:.1f}%"
    ])

headers = ["День недели (начало)", "Кол-во", "%", "Сумма НМЦК", "%"]
print(tabulate(table_data_notice, headers=headers, tablefmt="plain"))


max_count_idx, max_sum_idx = find_max_values(table_data_notice)
print(f"Наибольшее количество закупок: {table_data_notice[max_count_idx][0]} - {table_data_notice[max_count_idx][2]}")
print(f"Наибольшая сумма НМЦК: {table_data_notice[max_sum_idx][0]} - {table_data_notice[max_sum_idx][4]}")


top_days_table = dfw.groupby('date_notice').agg({
    'row_id': 'count',
    'nmck': 'sum'
}).sort_values('row_id', ascending=False).head(5)

top_days_table = top_days_table.reset_index()
top_days_table['date_notice'] = top_days_table['date_notice'].dt.strftime('%Y-%m-%d')
top_days_table['nmck'] = top_days_table['nmck'].apply(lambda x: format_number(x, 0))
top_days_table.columns = ['Дата публикации', 'Количество закупок', 'Сумма НМЦК']

print("\nТоп-5 дней с наибольшим количеством опубликованных закупок:")
print(tabulate(top_days_table, headers='keys', tablefmt='plain', showindex=False))

print("\nТоп-10 кодов ОКПД2 по количеству и по сумме:")


top_by_count = dfw['okpd2_group_code'].value_counts().head(10).index
top_by_sum = dfw.groupby('okpd2_group_code')['nmck'].sum().sort_values(ascending=False).head(10).index


all_top_codes = list(set(top_by_count) | set(top_by_sum))

table_data_comparison = []

for code in all_top_codes:
    count = (dfw['okpd2_group_code'] == code).sum()
    pct_count = (count / total_count) * 100
    code_sum = dfw[dfw['okpd2_group_code'] == code]['nmck'].sum()
    pct_sum = (code_sum / total_sum) * 100
    okpd_name = dfw[dfw['okpd2_group_code'] == code]['okpd2_group'].iloc[0]
    # Сокращаем название до 5 слов
    if okpd_name != "не указано":
        okpd_name = ' '.join(okpd_name.split()[:5])
    
    
    rank_count = list(top_by_count).index(code) + 1 if code in top_by_count else "-"
    rank_sum = list(top_by_sum).index(code) + 1 if code in top_by_sum else "-"
    
    table_data_comparison.append([
        code,
        okpd_name,
        format_number(count),
        f"{pct_count:.1f}%",
        rank_count,
        format_number(code_sum, 0),
        f"{pct_sum:.1f}%",
        rank_sum
    ])


table_data_comparison.sort(key=lambda x: int(x[4]) if x[4] != "-" else 999)

headers_comparison = ["Код ОКПД2", "Наименование", "Кол-во", "%", "Ранг кол-во", "Сумма НМЦК", "%", "Ранг сумма"]
print(tabulate(table_data_comparison, headers=headers_comparison, tablefmt="plain"))
```

##### Что касается влияния таких признаков, как НМЦК, месяцы и коды ОКПД2, это требует отдельного рассмотрения.

#### Начну с рассмотрения влияния НМЦК и сразу проверю несколько гипотиз:
Исходя из увиденного в предыдущих выводах хочу проверить несколько гипотез:
- Чем больше НМЦК, тем больше вероятность что будет подано 2+ заявок
- Чем меньше НМЦК, тем больше вероятность, что будет подана 1 заявка
- Чем меньше НМЦК, тем больше вероятность, что будет подано 0 заявок

Поскольку НМЦК имеет сильную ассиметрию, использую логарифмирование для того, чтобы распределение для каждой категории конкурентности (0, 1, 2+ заявок) было читаемым (выбросы сильно перетягивают распределение из-за чего не видно различий между категориями конкурентности). 
Статистические тесты выбраны исходя из типа распределения данных. Т.к. данные ассиметричны, применить параметрические тесты нельзя (они подходят для нармального распределения, а это не наш случай).
Для первой гипотез с учетом уже ранее написанного кода решаю использовать проверку гипотиз через анализ по квартилям (и для подверки верности выводов проверяю гипотизы ещё и логистической регрессией)

```python
print("\nПроверка гипотез двумя методами: логистическая регрессия и анализ по квартилям")


from sklearn.linear_model import LogisticRegression


X = dfw[['nmck']].values
X_log = np.log1p(X)  # Логарифмируем НМЦК для лучшей сходимости

# Гипотеза 1
y_2plus = dfw['is_two_plus'].values
model_2plus = LogisticRegression()
model_2plus.fit(X_log, y_2plus)
coef_2plus = model_2plus.coef_[0][0]

# Гипотеза 2 
y_1 = dfw['is_one'].values
model_1 = LogisticRegression()
model_1.fit(X_log, y_1)
coef_1 = model_1.coef_[0][0]

# Гипотеза 3
y_0 = dfw['is_zero'].values
model_0 = LogisticRegression()
model_0.fit(X_log, y_0)
coef_0 = model_0.coef_[0][0]

print("\nКоэффициенты логистической регрессии (влияние НМЦК на вероятность):")
print(f"Вероятность 2+ заявок: {coef_2plus:.4f}")
print(f"Вероятность 1 заявки: {coef_1:.4f}")
print(f"Вероятность 0 заявок: {coef_0:.4f}")


print("\nПроверка гипотез (логистическая регрессия):")

print("\nГипотеза 1: Чем больше НМЦК, тем больше вероятность 2+ заявок")
if coef_2plus > 0:
    print(f"Подтверждена: коэффициент положительный ({coef_2plus:.4f})")
    print("С ростом НМЦК вероятность 2+ заявок увеличивается")
else:
    print(f"Не подтверждена: коэффициент отрицательный ({coef_2plus:.4f})")

print("\nГипотеза 2: Чем меньше НМЦК, тем больше вероятность 1 заявки")
if coef_1 < 0:
    print(f"Подтверждена: коэффициент отрицательный ({coef_1:.4f})")
    print("С ростом НМЦК вероятность 1 заявки уменьшается")
else:
    print(f"Не подтверждена: коэффициент положительный ({coef_1:.4f})")

print("\nГипотеза 3: Чем меньше НМЦК, тем больше вероятность 0 заявок")
if coef_0 < 0:
    print(f"Подтверждена: коэффициент отрицательный ({coef_0:.4f})")
    print("С ростом НМЦК вероятность 0 заявок уменьшается")
else:
    print(f"Не подтверждена: коэффициент положительный ({coef_0:.4f})")



if 'nmck_quartile' not in dfw.columns:
    dfw['nmck_quartile'] = pd.qcut(dfw['nmck'], q=4, labels=['Q1 (низкий)', 'Q2', 'Q3', 'Q4 (высокий)'])


dfw['bids_category_label'] = dfw['bids_category'].map({0: '0', 1: '1', 2: '2+'}) # для наглядного отображения в анализе по квартилям


prob_analysis = dfw.groupby('nmck_quartile')['bids_category_label'].value_counts(normalize=True).unstack().fillna(0)

for category in ['0', '1', '2+']:
    if category not in prob_analysis.columns:
        prob_analysis[category] = 0

prob_analysis = prob_analysis[['0', '1', '2+']]


prob_0_q1 = prob_analysis.loc['Q1 (низкий)', '0']
prob_0_q4 = prob_analysis.loc['Q4 (высокий)', '0']

prob_1_q1 = prob_analysis.loc['Q1 (низкий)', '1']
prob_1_q4 = prob_analysis.loc['Q4 (высокий)', '1']

prob_2plus_q1 = prob_analysis.loc['Q1 (низкий)', '2+']
prob_2plus_q4 = prob_analysis.loc['Q4 (высокий)', '2+']

print("\nВероятности по квартилям НМЦК:")
print("Категория | Q1 (низкий) | Q4 (высокий) | Изменение")
print("-" * 55)
print(f"0 заявок  | {prob_0_q1:11.1%} | {prob_0_q4:12.1%} | {prob_0_q4 - prob_0_q1:+.1%}")
print(f"1 заявка  | {prob_1_q1:11.1%} | {prob_1_q4:12.1%} | {prob_1_q4 - prob_1_q1:+.1%}")
print(f"2+ заявок | {prob_2plus_q1:11.1%} | {prob_2plus_q4:12.1%} | {prob_2plus_q4 - prob_2plus_q1:+.1%}")

print("\nПроверка гипотез (анализ по квартилям):")

print("\nГипотеза 1: Чем больше НМЦК, тем больше вероятность 2+ заявок")
if prob_2plus_q4 > prob_2plus_q1:
    print(f"Подтверждена: вероятность растет с {prob_2plus_q1:.1%} до {prob_2plus_q4:.1%}")
    print(f"Рост: +{(prob_2plus_q4 - prob_2plus_q1)*100:.1f}%")
else:
    print(f"Не подтверждена: вероятность снижается с {prob_2plus_q1:.1%} до {prob_2plus_q4:.1%}")

print("\nГипотеза 2: Чем меньше НМЦК, тем больше вероятность 1 заявки")
if prob_1_q1 > prob_1_q4:
    print(f"Подтверждена: вероятность падает с {prob_1_q1:.1%} до {prob_1_q4:.1%}")
    print(f"Снижение: -{(prob_1_q1 - prob_1_q4)*100:.1f}%")
else:
    print(f"Не подтверждена: вероятность растет с {prob_1_q1:.1%} до {prob_1_q4:.1%}")

print("\nГипотеза 3: Чем меньше НМЦК, тем больше вероятность 0 заявок")
if prob_0_q1 > prob_0_q4:
    print(f"Подтверждена: вероятность падает с {prob_0_q1:.1%} до {prob_0_q4:.1%}")
    print(f"Снижение: -{(prob_0_q1 - prob_0_q4)*100:.1f}%")
else:
    print(f"Не подтверждена: вероятность растет с {prob_0_q1:.1%} до {prob_0_q4:.1%}")


print("\nСВОДНЫЙ РЕЗУЛЬТАТ:")

print("\nГипотеза 1: Чем больше НМЦК, тем больше вероятность 2+ заявок")
methods_1 = []
if coef_2plus > 0:
    methods_1.append("логистическая регрессия")
if prob_2plus_q4 > prob_2plus_q1:
    methods_1.append("анализ по квартилям")
    
if len(methods_1) == 2:
    print("Единогласно подтверждена обоими методами")
elif len(methods_1) == 1:
    print(f"Подтверждена только методом: {methods_1[0]}")
else:
    print("Не подтверждена ни одним методом")

print("\nГипотеза 2: Чем меньше НМЦК, тем больше вероятность 1 заявки")
methods_2 = []
if coef_1 < 0:
    methods_2.append("логистическая регрессия")
if prob_1_q1 > prob_1_q4:
    methods_2.append("анализ по квартилям")
    
if len(methods_2) == 2:
    print("Единогласно подтверждена обоими методами")
elif len(methods_2) == 1:
    print(f"Подтверждена только методом: {methods_2[0]}")
else:
    print("Не подтверждена ни одним методом")

print("\nГипотеза 3: Чем меньше НМЦК, тем больше вероятность 0 заявок")
methods_3 = []
if coef_0 < 0:
    methods_3.append("логистическая регрессия")
if prob_0_q1 > prob_0_q4:
    methods_3.append("анализ по квартилям")
    
if len(methods_3) == 2:
    print("Единогласно подтверждена обоими методами")
elif len(methods_3) == 1:
    print(f"Подтверждена только методом: {methods_3[0]}")
else:
    print("Не подтверждена ни одним методом")

# Сохранение результатов для последующего использования
nmck_analysis_results = {
    'logistic_coef_2plus': coef_2plus,
    'logistic_coef_1': coef_1,
    'logistic_coef_0': coef_0,
    'prob_2plus_q1': prob_2plus_q1,
    'prob_2plus_q4': prob_2plus_q4,
    'prob_1_q1': prob_1_q1,
    'prob_1_q4': prob_1_q4,
    'prob_0_q1': prob_0_q1,
    'prob_0_q4': prob_0_q4,
    'hypothesis_1_supported_methods': methods_1,
    'hypothesis_2_supported_methods': methods_2,
    'hypothesis_3_supported_methods': methods_3
}
```

Оба метода проверки гипотиз показали одинаковый результат. **Подтведилась только гипотиза № 3 "Чем меньше НМЦК, тем больше вероятность 0 заявок"**. В целом можно говорить о том, что исходя из проверки гипотиз мелкие закупки имеют высокие риски быть с 0 заявок, крупные закупки имеют высокую вероятность отсутствия конкуренции (1 заявка).

```python
print("Анализ влияния месяцев окончания подачи заявок на количество поданных заявок")

print("\nВлияние месяцев окончания подачи заявок на количество поданных заявок:")

month_deadline_stats = dfw.groupby('month_deadline').agg({
    'bids_cnt': ['count', 'mean', 'median'],
    'is_zero': 'mean',
    'is_two_plus': 'mean'
}).round(2)

month_deadline_stats.columns = ['Количество закупок', 'Среднее кол-во заявок', 'Медиана заявок', 
                               'Доля с 0 заявок', 'Доля с 2+ заявками']

month_order = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 
               'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
month_deadline_stats = month_deadline_stats.reindex(month_order)

print(month_deadline_stats)

best_month_deadline = month_deadline_stats['Доля с 2+ заявками'].idxmax()
worst_month_deadline = month_deadline_stats['Доля с 2+ заявками'].idxmin()
max_competition_deadline = month_deadline_stats['Доля с 2+ заявками'].max() * 100
min_competition_deadline = month_deadline_stats['Доля с 2+ заявками'].min() * 100

print(f"\nВыводы по месяцам окончания подачи:")
print(f"- Самый конкурентный месяц: {best_month_deadline} ({max_competition_deadline:.1f}% закупок с 2+ заявками)")
print(f"- Наименее конкурентный месяц: {worst_month_deadline} ({min_competition_deadline:.1f}% закупок с 2+ заявками)")
print(f"- Разница: {max_competition_deadline - min_competition_deadline:.1f} п.п.")

print("\nОбщие выводы по сезонности:\n")

def detect_seasons(month_stats):
    """Автоматически определяет высокий и низкий сезоны на основе данных"""
    
    sorted_months = month_stats.sort_values('Доля с 2+ заявками', ascending=False)
        
    high_season = sorted_months.head(4).index.tolist()
        
    low_season = sorted_months.tail(2).index.tolist()
    
    return high_season, low_season

high_season_deadline, low_season_deadline = detect_seasons(month_deadline_stats)
high_season_avg_deadline = month_deadline_stats.loc[high_season_deadline]['Доля с 2+ заявками'].mean() * 100
low_season_avg_deadline = month_deadline_stats.loc[low_season_deadline]['Доля с 2+ заявками'].mean() * 100

print(f"По месяцам окончания подачи:")
print(f"- Высокий сезон ({', '.join(high_season_deadline)}): {high_season_avg_deadline:.1f}% закупок с конкуренцией")
print(f"- Низкий сезон ({', '.join(low_season_deadline)}): {low_season_avg_deadline:.1f}% закупок с конкуренцией")
print(f"- Сезонная разница: {high_season_avg_deadline - low_season_avg_deadline:.1f} п.п.")

# Ищем 2 проблемных месяца с максимальной долей закупок без заявок
problem_months_deadline = month_deadline_stats.nlargest(2, 'Доля с 0 заявок')
problem_months_list = []
for month in problem_months_deadline.index:
    problem_rate = problem_months_deadline.loc[month, 'Доля с 0 заявок'] * 100
    problem_months_list.append(f"{month} ({problem_rate:.1f}%)")

print(f"\nПроблемные месяцы (максимальная доля закупок без заявок):")
for problem_month in problem_months_list:
    print(f"- {problem_month}")

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(month_deadline_stats.index, month_deadline_stats['Среднее кол-во заявок'], marker='o', linewidth=2, markersize=6)
plt.title('Среднее количество заявок по месяцам окончания подачи')
plt.xticks(rotation=45)
plt.ylabel('Среднее количество заявок')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
plt.plot(month_deadline_stats.index, month_deadline_stats['Доля с 0 заявок'], marker='o', color='red', linewidth=2, markersize=6)
plt.title('Доля закупок с 0 заявок по месяцам окончания подачи')
plt.xticks(rotation=45)
plt.ylabel('Доля закупок с 0 заявок')
plt.grid(True, alpha=0.3)

# Подсвечиваем проблемные месяцы на графике
problem_indices = [month_deadline_stats.index.get_loc(month) for month in problem_months_deadline.index]
problem_values = [month_deadline_stats.loc[month, 'Доля с 0 заявок'] for month in problem_months_deadline.index]
plt.scatter(problem_indices, problem_values, color='darkred', s=100, zorder=5, label='Проблемные месяцы')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(month_deadline_stats.index, month_deadline_stats['Доля с 2+ заявками'], marker='o', color='green', linewidth=2, markersize=6)
plt.title('Доля закупок с 2+ заявками по месяцам окончания подачи')
plt.xticks(rotation=45)
plt.ylabel('Доля закупок с 2+ заявками')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 4)
plt.bar(month_deadline_stats.index, month_deadline_stats['Количество закупок'], alpha=0.7)
plt.title('Количество закупок по месяцам окончания подачи')
plt.xticks(rotation=45)
plt.ylabel('Количество закупок')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(FIG / 'month_analysis_deadline_only.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("Анализ ОКПД2\n")

print("Уникальные коды ОКПД2 в данных:")
total_unique_codes = dfw['okpd2_group_code'].nunique()
print(f"Всего уникальных кодов ОКПД2: {total_unique_codes}")

def format_money(x):
    if x >= 1e9:
        return f'{x/1e9:.1f} млрд'
    elif x >= 1e6:
        return f'{x/1e6:.1f} млн'
    else:
        return f'{x/1e3:.0f} тыс'

print("\nТоп-10 ОКПД2 по объему НМЦК в разрезе успешности закупок:")

# 2.1 Топ-10 по объему НМЦК в разрезе успешности
for i, success_cat in enumerate(['is_zero', 'is_one', 'is_two_plus']):
    # Фильтруем данные по категории успешности
    mask = dfw[success_cat] == 1
    temp_df = dfw[mask]
    
    top_nmck = temp_df.groupby('okpd2_group_code').agg({
        'nmck': 'sum',
        'okpd2_group': 'first',
        'row_id': 'count'
    }).nlargest(10, 'nmck')
    
    # Выводим таблицу в консоль
    print(f"\n{['Закупки с 0 заявок', 'Закупки с 1 заявкой', 'Закупки с 2+ заявками'][i]}:")
    display_data = top_nmck.reset_index()
    display_data['nmck_formatted'] = display_data['nmck'].apply(format_money)
    display_data['count'] = display_data['row_id']
    print(tabulate(display_data[['okpd2_group_code', 'okpd2_group', 'count', 'nmck_formatted']], 
                   headers=['Код', 'Название', 'Кол-во', 'Сумма НМЦК'], 
                   tablefmt='pretty', showindex=False))

print("\nТоп-10 ОКПД2 по количеству закупок в разрезе успешности:")

for i, success_cat in enumerate(['is_zero', 'is_one', 'is_two_plus']):
    # Фильтруем данные по категории успешности
    mask = dfw[success_cat] == 1
    temp_df = dfw[mask]
    
    top_count = temp_df.groupby('okpd2_group_code').agg({
        'okpd2_group': 'first',
        'row_id': 'count'
    }).nlargest(10, 'row_id')
    
    print(f"\n{['Закупки с 0 заявок', 'Закупки с 1 заявкой', 'Закупки с 2+ заявками'][i]}:")
    display_data = top_count.reset_index()
    display_data.columns = ['okpd2_group_code', 'okpd2_group', 'count']
    print(tabulate(display_data[['okpd2_group_code', 'okpd2_group', 'count']], 
                   headers=['Код', 'Название', 'Кол-во закупок'], 
                   tablefmt='pretty', showindex=False))

print("\nТоп-10 ОКПД2 по способам закупок:")

proc_methods = dfw['proc_method'].unique()
print(f"Найдены способы закупок: {list(proc_methods)}")

for method in proc_methods:
    temp_df = dfw[dfw['proc_method'] == method]
    
    top_nmck_method = temp_df.groupby('okpd2_group_code').agg({
        'nmck': 'sum',
        'okpd2_group': 'first',
        'row_id': 'count'
    }).nlargest(10, 'nmck')
    
    print(f"\n{method} - Топ-10 по объему НМЦК:")
    display_data_method = top_nmck_method.reset_index()
    display_data_method['nmck_formatted'] = display_data_method['nmck'].apply(format_money)
    display_data_method['count'] = display_data_method['row_id']
    print(tabulate(display_data_method[['okpd2_group_code', 'okpd2_group', 'count', 'nmck_formatted']], 
                   headers=['Код', 'Название', 'Кол-во', 'Сумма НМЦК'], 
                   tablefmt='pretty', showindex=False))
    
    top_count_method = temp_df.groupby('okpd2_group_code').agg({
        'okpd2_group': 'first',
        'row_id': 'count'
    }).nlargest(10, 'row_id')
    
    print(f"\n{method} - Топ-10 по количеству закупок:")
    display_count_method = top_count_method.reset_index()
    display_count_method.columns = ['okpd2_group_code', 'okpd2_group', 'count']
    print(tabulate(display_count_method[['okpd2_group_code', 'okpd2_group', 'count']], 
                   headers=['Код', 'Название', 'Кол-во закупок'], 
                   tablefmt='pretty', showindex=False))

print("\nСводная статистика ОКПД2:")
print(f"Всего уникальных кодов ОКПД2: {total_unique_codes}")
print(f"Общее количество закупок: {len(dfw)}")
print(f"Среднее количество закупок на один код ОКПД2: {len(dfw)/total_unique_codes:.1f}")

# ДОБАВЛЕННЫЙ КОД: Создание okpd2_stats перед использованием
print(f"\nРаспределение закупок по кодам ОКПД2:")
okpd2_stats = dfw['okpd2_group_code'].value_counts().reset_index()
okpd2_stats.columns = ['okpd2_group_code', 'count']
okpd2_stats['percentage'] = (okpd2_stats['count'] / len(dfw) * 100).round(3)
okpd2_stats['cumulative_percentage'] = okpd2_stats['percentage'].cumsum()

top_10_codes_coverage = okpd2_stats.head(10)['percentage'].sum()
top_20_codes_coverage = okpd2_stats.head(20)['percentage'].sum()
print(f"Топ-10 кодов покрывают: {top_10_codes_coverage:.1f}% всех закупок")
print(f"Топ-20 кодов покрывают: {top_20_codes_coverage:.1f}% всех закупок")

print(f"\nАнализ покрытия закупок:")
okpd2_coverage = dfw['okpd2_group_code'].value_counts().reset_index()
okpd2_coverage.columns = ['okpd2_group_code', 'count']
okpd2_coverage['percentage'] = (okpd2_coverage['count'] / len(dfw) * 100).round(3)
okpd2_coverage['cumulative_percentage'] = okpd2_coverage['percentage'].cumsum()

target_coverage = 95
codes_for_95 = okpd2_coverage[okpd2_coverage['cumulative_percentage'] >= target_coverage].iloc[0]
num_codes_95 = okpd2_coverage[okpd2_coverage['cumulative_percentage'] <= target_coverage].shape[0] + 1

print(f"Для покрытия {target_coverage}% всех закупок нужно: {num_codes_95} кодов ОКПД2")
print(f"Это {num_codes_95}/{len(okpd2_coverage)} кодов ({num_codes_95/len(okpd2_coverage)*100:.1f}% от всех кодов)")

print(f"\nАнализ покрытия закупок:")
coverage_levels = [50, 80, 90, 95, 99]
for level in coverage_levels:
    if level <= okpd2_coverage['cumulative_percentage'].max():
        codes_needed = okpd2_coverage[okpd2_coverage['cumulative_percentage'] <= level].shape[0] + 1
        actual_coverage = okpd2_coverage[okpd2_coverage['cumulative_percentage'] <= level]['cumulative_percentage'].max()
        print(f"{level}% закупок покрывают: {codes_needed:3d} кодов (фактически {actual_coverage:.1f}%)")

tail_codes = len(okpd2_coverage) - num_codes_95
tail_coverage = 100 - okpd2_coverage.iloc[num_codes_95 - 1]['cumulative_percentage']
print(f"\nАнализ распределения:")
print(f"Топ-{num_codes_95} кодов: покрывают {okpd2_coverage.iloc[num_codes_95 - 1]['cumulative_percentage']:.1f}% закупок")
print(f"Остальные {tail_codes} кодов: покрывают всего {tail_coverage:.1f}% закупок")
print(f"Среднее количество закупок на код в 'хвосте': {okpd2_coverage.iloc[num_codes_95:]['count'].mean():.1f}")

detailed_okpd2_stats = dfw.groupby('okpd2_group_code').agg({
    'okpd2_group': 'first',
    'row_id': 'count',
    'nmck': ['sum', 'mean', 'median'],
    'is_zero': 'mean',
    'is_one': 'mean',
    'is_two_plus': 'mean',
    'proc_method': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'unknown'
}).round(3)

detailed_okpd2_stats.columns = ['name', 'count', 'nmck_sum', 'nmck_mean', 'nmck_median', 
                               'zero_rate', 'one_rate', 'competitive_rate', 'most_common_method']
detailed_okpd2_stats = detailed_okpd2_stats.sort_values('count', ascending=False)

detailed_okpd2_stats.to_csv(CSV / 'detailed_okpd2_statistics.csv', encoding='utf-8-sig')
```

Также, прежде чем перейти к дальнейшему анализу нужно создать категоризированную колонку для НМЦК (опираясь на предыдущий анализ, где было деление на 4 квартиля, приняла решение что для повышения точности необходимо разделить на 5 квартилей (так сохраниться баланс между хорошей детализацией и умеренным количеством категорий)

```python
print("Категориальный признак для НМЦК (5 квартилей)\n")

quantiles = [0.2, 0.4, 0.6, 0.8]
q20 = dfw['nmck'].quantile(0.2)
q40 = dfw['nmck'].quantile(0.4)
q60 = dfw['nmck'].quantile(0.6)
q80 = dfw['nmck'].quantile(0.8)

print("Границы квартилей НМЦК:")
print(f"20% перцентиль: {format_number(q20)} руб.")
print(f"40% перцентиль: {format_number(q40)} руб.")
print(f"60% перцентиль: {format_number(q60)} руб.")
print(f"80% перцентиль: {format_number(q80)} руб.")

# Создаем новую колонку с 5 категориями
def categorize_nmck_5_quantiles(value):
    """Разделяет НМЦК на 5 категорий по квартилям"""
    if value <= q20:
        return 'q1_мелкие'
    elif value <= q40:
        return 'q2_маленькие'
    elif value <= q60:
        return 'q3_средние'
    elif value <= q80:
        return 'q4_крупные'
    else:
        return 'q5_очень_крупные'

dfw['nmck_5_quantiles'] = dfw['nmck'].apply(categorize_nmck_5_quantiles)

print("\nСоздан новый признак НМЦК для дальнейшего анализа 'nmck_5_quantiles'")
print(f"Категории: {dfw['nmck_5_quantiles'].unique().tolist()}")
```

#### Определение влияния признаков на успешность закупки

Поскольку данные имеют правосторонний перекос, использование стандартных параметрических методов (как корреляция Пирсона) не рекомендуется, будем использовать другие методы (и сравним их в итоге)

```python
from scipy.stats import pearsonr, spearmanr, kruskal, mannwhitneyu, chi2_contingency

print("статистический анализ с учетом правостороннего перекоса")

temp_df = dfw[['is_successful', 'okpd2_optimized', 'nmck_5_quantiles', 'proc_method', 'nmck']].copy()

nmck_encoding = {
    'q1_мелкие': 1,
    'q2_маленькие': 2, 
    'q3_средние': 3,
    'q4_крупные': 4,
    'q5_очень_крупные': 5
}
temp_df['nmck_encoded'] = temp_df['nmck_5_quantiles'].map(nmck_encoding)


print("\nкорреляция пирсона:")
pearson_okpd2, _ = pearsonr(temp_df['is_successful'], temp_df['nmck_encoded'])
pearson_method, _ = pearsonr(temp_df['is_successful'], pd.factorize(dfw['proc_method'])[0])

print("окпд2:")
print(f"корреляция пирсона: {pearson_okpd2:.3f}")

print("\nподтверждение выводов альтернативными методами")

print("корреляция спирмена:")
spearman_corr_nmck, p_value_spearman = spearmanr(temp_df['is_successful'], temp_df['nmck_encoded'])
print(f"размер нмцк: {spearman_corr_nmck:.3f} (p-value: {p_value_spearman:.4f})")

print("\nкритерий краскела-уоллиса:")

okpd2_groups = [group['is_successful'].values for name, group in dfw.groupby('okpd2_optimized')]
h_stat_okpd2, p_value_kw_okpd2 = kruskal(*okpd2_groups)
print(f"окпд2: h={h_stat_okpd2:.1f}, p-value={p_value_kw_okpd2:.4f}")

method_groups = [group['is_successful'].values for name, group in dfw.groupby('proc_method')]
h_stat_method, p_value_kw_method = kruskal(*method_groups)
print(f"способ закупки: h={h_stat_method:.1f}, p-value={p_value_kw_method:.4f}")

print("\nтест манна-уитни для парных сравнений:")

successful_nmck = dfw[dfw['is_successful'] == 1]['nmck']
unsuccessful_nmck = dfw[dfw['is_successful'] == 0]['nmck']

u_stat_nmck, p_value_mw_nmck = mannwhitneyu(
    np.log1p(successful_nmck), 
    np.log1p(unsuccessful_nmck)
)
print(f"различия в нмцк: u={u_stat_nmck:.0f}, p-value={p_value_mw_nmck:.4f}")

print("\nтест хи-квадрат:")

contingency_method = pd.crosstab(dfw['proc_method'], dfw['is_successful'])
chi2_method, p_chi2_method, dof_method, expected_method = chi2_contingency(contingency_method)
print(f"способ закупки: χ²={chi2_method:.1f}, p-value={p_chi2_method:.4f}")

top_okpd2 = dfw['okpd2_optimized'].value_counts().head(15).index
contingency_okpd2 = pd.crosstab(
    dfw[dfw['okpd2_optimized'].isin(top_okpd2)]['okpd2_optimized'], 
    dfw[dfw['okpd2_optimized'].isin(top_okpd2)]['is_successful']
)
chi2_okpd2, p_chi2_okpd2, dof_okpd2, expected_okpd2 = chi2_contingency(contingency_okpd2)
print(f"окпд2 (топ-15): χ²={chi2_okpd2:.1f}, p-value={p_chi2_okpd2:.4f}")

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

print("\nкоэффициент v крамера (сила связи):")
v_okpd2 = cramers_v(dfw['okpd2_optimized'], dfw['is_successful'])
v_method = cramers_v(dfw['proc_method'], dfw['is_successful'])
print(f"ОКПД2: {v_okpd2:.3f}")
print(f"способ закупки: {v_method:.3f}")

print("\nитоговая интерпретация с учетом перекоса данных:")

print("окпд2:")
print(f"   • критерий краскела-уоллиса p-value: {p_value_kw_okpd2:.4f} {'статистически значимо' if p_value_kw_okpd2 < 0.05 else 'не значимо'}")
print(f"   • хи-квадрат p-value: {p_chi2_okpd2:.4f} {'статистически значимо' if p_chi2_okpd2 < 0.05 else 'не значимо'}")
print(f"   • v крамера: {v_okpd2:.3f}")

print("\nразмер нмцк:")
print(f"   • корреляция спирмена: {spearman_corr_nmck:.3f} (p-value: {p_value_spearman:.4f})")
print(f"   • тест манна-уитни p-value: {p_value_mw_nmck:.4f} {'статистически значимо' if p_value_mw_nmck < 0.05 else 'не значимо'}")

print("\nспособ закупки:")
print(f"   • критерий краскела-уоллиса p-value: {p_value_kw_method:.4f} {'статистически значимо' if p_value_kw_method < 0.05 else 'не значимо'}")
print(f"   • хи-квадрат p-value: {p_chi2_method:.4f} {'статистически значимо' if p_chi2_method < 0.05 else 'не значимо'}")
print(f"   • v крамера: {v_method:.3f}")

print("\nключевые выводы:")
print("_"*60)

if p_value_kw_okpd2 < 0.05 and p_chi2_okpd2 < 0.05:
    print("ОКПД2: высокая статистическая значимость (подтверждено непараметрическими тестами)")
else:
    print("ОКПД2: статистическая значимость под вопросом")

if p_value_spearman < 0.05 or p_value_mw_nmck < 0.05:
    print("размер нмцк: статистически значим (подтверждено непараметрическими тестами)")
else:
    print("размер нмцк: не значим")

if p_value_kw_method < 0.05 and p_chi2_method < 0.05:
    print("способ закупки: статистически значим (подтверждено непараметрическими тестами)")
else:
    print("способ закупки: не значим")
```

Выводы можно считать надежными, поскольку они подтверждены (и соответственно подверенны, с использованием непараметрических методов (решила оставить и "Пирсона", хотя и понимаю, что для данных с перекосом он не годится. Это был умышленный шаг в попытке увидеть как можно больше).

##### После того, как увидели, что на первом месте по значимости стоит признак "ОКПД2", посмотрим какие коды ОКПД2 наиболее часто остаются без заявок или только с 1 заявкой. Для этого сделаем 2 топа (15).

*Определяем топ сразу по двум показателям (и по количеству и по сумме), отбираем топ ОКПД2 в которых оба ранга (по количеству и по сумме) занимают высокую позицию.*

```python
df_all_years = dfw[dfw['year_dl'].isin([2023, 2024, 2025])]

print("ПРОБЛЕМНЫЕ ОКПД2")

def shorten_name(name, max_words=5):
    if isinstance(name, str):
        words = name.split()
        if len(words) > max_words:
            return ' '.join(words[:max_words]) + '...'
        else:
            return name
    else:
        return "не указано"

# Закупки без заявок
zero_all = df_all_years[df_all_years['is_zero'] == 1].groupby('okpd2_group_code').agg({
    'nmck': ['sum', 'count', 'median'],
    'okpd2_group': 'first'
}).round(2)
zero_all.columns = ['Совокупная_НМЦК', 'Количество_закупок', 'Медианная_НМЦК', 'Наименование']


zero_all['Ранг_по_сумме'] = zero_all['Совокупная_НМЦК'].rank(ascending=False, method='min').astype(int)
zero_all['Ранг_по_количеству'] = zero_all['Количество_закупок'].rank(ascending=False, method='min').astype(int)
zero_all['Средний_ранг'] = (zero_all['Ранг_по_сумме'] + zero_all['Ранг_по_количеству']) / 2


zero_combined = zero_all.sort_values('Средний_ранг').head(15)

print("\nТОП-15 ОКПД2 БЕЗ ЗАЯВОК С ВЫСОКИМИ РАНГАМИ ПО ОБОИМ ПОКАЗАТЕЛЯМ")
print("(ранги рассчитаны среди всех закупок без заявок за 2023-2025 годы)")
table_data = []
for i, (okpd2_code, row) in enumerate(zero_combined.iterrows(), 1):
    table_data.append([
        i,
        okpd2_code,
        shorten_name(row['Наименование']),
        format_number(row['Совокупная_НМЦК'], 0),
        int(row['Количество_закупок']),
        format_number(row['Медианная_НМЦК'], 0),
        row['Ранг_по_сумме'],
        row['Ранг_по_количеству'],
        round(row['Средний_ранг'], 1)
    ])

headers = ["№", "ОКПД2", "Наименование", "Сумма", "Кол-во", "Медиана", "Ранг сумм.", "Ранг кол-во", "Ср.ранг"]
print(tabulate(table_data, headers=headers, tablefmt="plain", stralign="left", numalign="left"))

# Закупки с одной заявкой
one_all = df_all_years[df_all_years['is_one'] == 1].groupby('okpd2_group_code').agg({
    'nmck': ['sum', 'count', 'median'],
    'okpd2_group': 'first'
}).round(2)
one_all.columns = ['Совокупная_НМЦК', 'Количество_закупок', 'Медианная_НМЦК', 'Наименование']

# Добавляем ранги (среди всех закупок с одной заявкой)
one_all['Ранг_по_сумме'] = one_all['Совокупная_НМЦК'].rank(ascending=False, method='min').astype(int)
one_all['Ранг_по_количеству'] = one_all['Количество_закупок'].rank(ascending=False, method='min').astype(int)
one_all['Средний_ранг'] = (one_all['Ранг_по_сумме'] + one_all['Ранг_по_количеству']) / 2

# Отбираем ТОП-15 по среднему рангу
one_combined = one_all.sort_values('Средний_ранг').head(15)

print("\nТОП-15 ОКПД2 С ОДНОЙ ЗАЯВКОЙ С ВЫСОКИМИ РАНГАМИ ПО ОБОИМ ПОКАЗАТЕЛЯМ")
print("(ранги рассчитаны среди всех закупок с одной заявкой за 2023-2025 годы)")
table_data = []
for i, (okpd2_code, row) in enumerate(one_combined.iterrows(), 1):
    table_data.append([
        i,
        okpd2_code,
        shorten_name(row['Наименование']),
        format_number(row['Совокупная_НМЦК'], 0),
        int(row['Количество_закупок']),
        format_number(row['Медианная_НМЦК'], 0),
        row['Ранг_по_сумме'],
        row['Ранг_по_количеству'],
        round(row['Средний_ранг'], 1)
    ])

headers = ["№", "ОКПД2", "Наименование", "Сумма", "Кол-во", "Медиана", "Ранг сумм.", "Ранг кол-во", "Ср.ранг"]
print(tabulate(table_data, headers=headers, tablefmt="plain", stralign="left", numalign="left"))

print(f"\n*Всего закупок без заявок: {len(zero_all)} ОКПД2 категорий")
print(f"*Всего закупок с одной заявкой: {len(one_all)} ОКПД2 категорий")
print("*Сортировка по среднему рангу выявляет категории с высокими рангами как по сумме, так и по количеству")
```

При изучении данных было выявлено, что в топе с 0 заявок нет ОКПД2 который на 3 месте по сумме, решила отдельно посмотреть что это за ОКПД2 и почему он не попал в топ.

```python
third_rank_by_sum = zero_all[zero_all['Ранг_по_сумме'] == 3]

if len(third_rank_by_sum) > 0:
    print("\nОКПД2 с 3 рангом по сумме (среди всех закупок без заявок):\n")
    
    for okpd2_code, row in third_rank_by_sum.iterrows():
        print(f"ОКПД2 код: {okpd2_code}")
        print(f"Наименование: {row['Наименование']}")
        print(f"Совокупная сумма: {format_number(row['Совокупная_НМЦК'], 0)}")
        print(f"Количество закупок: {int(row['Количество_закупок'])}")
        print(f"Медианная НМЦК: {format_number(row['Медианная_НМЦК'], 0)}")
        print(f"Ранг по сумме: {row['Ранг_по_сумме']}")
        print(f"Ранг по количеству: {row['Ранг_по_количеству']}")
        print(f"Средний ранг: {round(row['Средний_ранг'], 1)}")

        print(f"\nпочему не попал в топ:")
        top_threshold = zero_combined['Средний_ранг'].max()
        print(f"Порог попадания в топ-15: {top_threshold}")
        print(f"Попадает в топ? {'ДА' if row['Средний_ранг'] <= top_threshold else 'НЕТ'}")
```

#### Опишем портрет среднестатистической закупки по уровням конкуренции:
* ##### не состоявшаяся (0 заявок)
* ##### без конкуренции (1 заявка)
* ##### с конкуренцией (2+ заявок)

```python
print("Среднестатистический портрет закупки\n")

def create_portrait(bid_type, label):
    """Создает портрет закупки для заданного типа"""
    subset = dfw[dfw[bid_type] == 1]
    
    if len(subset) == 0:
        return ["Нет данных"] * 5
    
    
    most_common_method = subset['proc_method'].mode()
    method = most_common_method[0] if not most_common_method.empty else "нет данных"
    
    
    median_nmck = subset['nmck'].median()
    
    
    most_common_okpd = subset['okpd2_group_code'].mode()
    okpd_code = most_common_okpd[0] if not most_common_okpd.empty else "нет данных"
    okpd_name = subset.loc[subset['okpd2_group_code'] == okpd_code, 'okpd2_group'].iloc[0] if not most_common_okpd.empty else "нет данных"
    
    
    most_common_dow = subset['dow_deadline'].mode()
    dow = most_common_dow[0] if not most_common_dow.empty else "нет данных"
    short_dow = {'понедельник': 'ПН', 'вторник': 'ВТ', 'среда': 'СР', 'четверг': 'ЧТ', 
                 'пятница': 'ПТ', 'суббота': 'СБ', 'воскресенье': 'ВС'}.get(dow, dow)
    
    
    most_common_month = subset['month_num_dl'].mode()
    month_num = most_common_month[0] if not most_common_month.empty else "нет данных"
    month_names = {1: 'январе', 2: 'феврале', 3: 'марте', 4: 'апреле', 5: 'мае', 6: 'июне',
                   7: 'июле', 8: 'августе', 9: 'сентябре', 10: 'октябре', 11: 'ноябре', 12: 'декабре'}
    month = month_names.get(month_num, 'нет данных')
    
    return [method, f"{format_number(median_nmck, 0)}", okpd_code, okpd_name, short_dow, month]


for bid_type, label in [('is_zero', '0 заявок'), 
                        ('is_one', '1 заявкой'), 
                        ('is_two_plus', '2+ заявками')]:
    
    subset = dfw[dfw[bid_type] == 1]
    
    if len(subset) > 0:
        method = subset['proc_method'].mode()[0]
        median_nmck = format_number(subset['nmck'].median(), 0)
        okpd_code = subset['okpd2_group_code'].mode()[0]
        okpd_name = subset.loc[subset['okpd2_group_code'] == okpd_code, 'okpd2_group'].iloc[0]
        dow = subset['dow_deadline'].mode()[0]
        short_dow = {'понедельник': 'понедельник', 'вторник': 'вторник', 'среда': 'среду', 'четверг': 'четверг', 
                     'пятница': 'пятницу', 'суббота': 'субботу', 'воскресенье': 'воскресенье'}.get(dow, dow)
        
        print(f"Среднестатистическая закупка с {label}:")
        print(f"   • Это {method.lower()} с медианной НМЦК {median_nmck} руб.")
        print(f"   • Код ОКПД2: {okpd_code} ({okpd_name})")
        print(f"   • Чаще всего проходит во {short_dow}")
        print()
```
